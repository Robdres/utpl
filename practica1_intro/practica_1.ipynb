```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
database = pd.read_csv("./diabetes_012_health_indicators_BRFSS2015.csv")
#You can find the database here https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset/data
```

```python
database.head()
```

```python
#Classificacion para cada objecto
classification = database["Diabetes_012"]
classification.head()
```

```python
#Atributos sin la classificacion
features =  database[database.columns[1:]]
features.head()
```

```python
#Estos son los atributos con los que trabajaremos
print(features.columns)
print(len(features.columns)) #originalmente tenemos 21 atributos
```

```python
#Aumentaremos un par de columnas para practicar
# Estos son los rangos para edad del dataset
#1 Age 18 to 24
#2 Age 25 to 29
#3 Age 30 to 34
#4 Age 35 to 39
#5 Age 40 to 44
#6 Age 45 to 49
#7 Age 50 to 54
#8 Age 55 to 59
#9 Age 60 to 64
#10 Age 65 to 69
#11 Age 70 to 74
#12 Age 75 to 79
#13 Age 80 or older

#1. Agregar columna de personas mayores de 60

elder = (database["Age"]>=9).rename("elder")
extra_1_feature = features.join(elder)

print(extra_1_feature.columns)
print(len(extra_1_feature.columns))# Aqui tenemos 22
```

```python
extra_1_feature.head()
```

```python
# Aqui podemos ver que existen alrededor de 12000 personas mayores de 60 en este estudio
extra_1_feature["elder"].value_counts()
```

```python
# Aqui un pequeño ejemplo para simplificar como se pueden hacer un summary de columnas
# 2. Agregar una columna para considerar frutas y verduras lo mismo, (solo por practicar) 
# Aqui se convierte en dos categorias, 0 no come saludable, 1 come saludable, 2 come muy saludable

fruit_or_vegetable = extra_1_feature["Fruits"] + extra_1_feature["Veggies"]
extra_2_feature = extra_1_feature
extra_2_feature = extra_2_feature.join(fruit_or_vegetable.rename("healthy_alimentation"))
extra_2_feature.head()
```

```python
print(extra_2_feature.columns)
print(len(extra_2_feature.columns)) # ahora tenemos 23
```

```python
# Finalmente voy a aumentar dos para no perder el tiempo explicando cada una. Una columna es el bmi standarisado,
#y voy a hacer una columna menth health weekly, ya que esta columna dice los días que no te sentiste al 100% mentalmente en un 
#mes de 30 diás

bmi = features["BMI"]
bmi = (bmi-bmi.min())/(bmi.max()-bmi.min())
bmi
```

```python
mental_weeks = features["MentHlth"]
mental_weeks = mental_weeks/7
mental_weeks
```

```python
features_4_extra_columns = extra_2_feature
features_4_extra_columns= features_4_extra_columns.join([bmi.rename("std_bmi"),mental_weeks.rename("mental_week")])
features_4_extra_columns.head()
```

```python
print(features_4_extra_columns.columns)
print(len(features_4_extra_columns.columns)) # ahora tenemos 25 columnas
```

```python
# Aqui esta el "nuevo" database con 4 columnas extra
features_2 = features_4_extra_columns
features_2 
```

```python
# Voy a visualizar algunos de los datos
# Primero la distribucion de datos
age = features_2["Age"]
print(age.value_counts().sort_index())
indexes = age.unique()
indexes.sort()
print(indexes)
```

```python
# Mapear la edad con el BMI
subset = features_2.sample(5000)
subset = subset.sort_values("Age")
age = subset["Age"]
bmi = subset["BMI"]
plt.plot( age, bmi, linestyle='None', marker="o" )
plt.xticks(rotation="vertical")
plt.title("Edad vs BMI")
plt.show()
# esto por ejemplo nos dice que parece que hay una distribución buena entre edades y bmi. 
```

```python
#Ahora vamos a hacer dos gráficos para entender la relación entre los valores y la clasificación de diabétes
# Mapear la edad con la diabetes
diabetes = database[database["Diabetes_012"]==2]
pre_diabetes = database[database["Diabetes_012"]==1]
no_diabetes = database[database["Diabetes_012"]==0]
fig,ax = plt.subplots(1,3,figsize=(12,4))
ax[0].hist( diabetes["Age"] )
ax[0].set_title("Edad de diabeticos")
ax[1].hist( pre_diabetes["Age"] )
ax[1].set_title("Edad de prediabeticos")
ax[2].hist( no_diabetes["Age"] )
ax[2].set_title("Edad de no diabeticos")
plt.show()
# podemos ver la relación de edad y los datos
```

```python
#Para el análisis de los datos podemos hacer uso de una red multi-capa de perceptrones
from sklearn.model_selection import train_test_split
#primero dividimos en training y testing data
training, test = train_test_split( database, test_size=0.2)
```

```python
print("Training data size " + str(len(training)))
print("Testing data size " + str(len(test)))
```

```python
#Dividimos cada uno de ellos en features y classes
features_training, classes_training =  training[training.columns[1:]],training["Diabetes_012"].to_frame()
features_test, classes_test =  test[test.columns[1:]], test["Diabetes_012"].to_list()
```

```python
#Para la clasificación vamos a usar un RandomForest Classifier, basado en un buen resultado de 
#https://www.kaggle.com/code/thanayodchira/diabetes. (El trabajo no es para ver diferentes modelos asi que elegiremos uno que funciona)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=200)
# Aqui entrenamos el modelo
model.fit(features_training,classes_training)
```

```python
# Ahora que tenemos nuestro modelo entrenado, vamos a predecir los valores de nuesta clase de prueba
results = model.predict(features_test)
print(results)
```

```python
# Tuvimos buenos resultados, accuracy alredor de 0.84
from sklearn.metrics import  classification_report
print(classification_report(classes_test, results))
```
